```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df = pd.read_csv("titanic/tested.csv")
df.head()
```

```python
#checking for null values
```

```python
df.isnull().sum()
```

```python
#dropping Cabin column because it has lots of null value
```

```python
df.drop("Cabin",axis=1,inplace = True)
```

```python
df
```

```python
#calculating mean of Age column
```

```python
agemean = df["Age"].mean()
```

```python
agemean
```

```python
#filling null values inside Age column with Age mean values
```

```python
df["Age"].fillna(agemean,inplace=True)
```

```python
df
```

```python
#replacing male as 1 and female as 0 and also embarked column
```

```python
df.replace({"Sex":{"female":0,"male":1},"Embarked":{"S":0,"C":1,"Q":2}},inplace=True)
```

```python
df
```

```python
#checking again if any null value is pending
```

```python
df.info()
```

```python
#dataset is now clean to fit into model
```

```python
#collecting all the independent variables into X (Pclass,Sex,Age,Embarked)
#collection dependent variable or target variable into Y (Survived)
```

```python
X = df[["Pclass","Age","Sex","Embarked"]]
Y = df["Survived"]
```

```python
X
```

```python
Y
```

```python
#splitting data into train part and test part
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
#creating a model of LogisticRegression
```

```python
L_model = LogisticRegression()
```

```python
L_model.fit(X_train,Y_train)
```

```python
predicted_value_of_training = L_model.predict(X_train)
```

```python
predicted_value_of_training
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of model in terms of train dataset is  ",accuracy_score(Y_train,predicted_value_of_training))
```

```python
#lets check the accuracy with test dataset
```

```python
predicted_value_of_test = L_model.predict(X_test)
```

```python
predicted_value_of_test
```

```python
print("Accuracy of model in terms of test dataset is ",accuracy_score(Y_test,predicted_value_of_test))
```
